```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
CNNTorch = pd.read_csv("./results/torch/CNNTorch-result.csv")
LeNetTorch = pd.read_csv("./results/torch/LeNetTorch-result.csv")
LeNetProTorch = pd.read_csv("./results/torch/LeNetProTorch-result.csv")
```

```python
CNNTorch.head()
```

```python
# 设置线条的样式
sns.set_style("whitegrid")
# 设置绘图的风格
# sns.set_context("poster")
# 设置绘图的尺寸
# plt.figure(figsize=(10, 5))
# 设置线条的宽度
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
plt.plot(CNNTorch['train_score'][:100],label="cnn train score")
plt.plot(LeNetTorch['train_score'],label="lenet train score")
plt.plot(LeNetProTorch['train_score'],label="lenetpro train score")
plt.legend()
plt.title("train score")
plt.savefig("./results/torch/train_score.png")
```

```python
# 设置线条的样式
sns.set_style("whitegrid")
# 设置绘图的风格
# sns.set_context("poster")
# 设置绘图的尺寸
# plt.figure(figsize=(10, 5))
# 设置线条的宽度
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
plt.plot(CNNTorch['test_score'][:100],label="cnn test score")
plt.plot(LeNetTorch['test_score'],label="lenet test score")
plt.plot(LeNetProTorch['test_score'],label="lenetpro test score")
plt.legend()
plt.title("test score")
plt.savefig("./results/torch/test_score.png")
```

```python
# 设置线条的样式
sns.set_style("whitegrid")
# 设置绘图的风格
# sns.set_context("poster")
# 设置绘图的尺寸
# plt.figure(figsize=(10, 5))
# 设置线条的宽度
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
plt.plot(CNNTorch['epoch_loss'][:100],label="cnn loss")
plt.plot(LeNetTorch['epoch_loss'],label="lenet loss")
plt.plot(LeNetProTorch['epoch_loss'],label="lenetpro loss")
plt.legend()
plt.title("loss")
plt.savefig("./results/torch/loss.png")
```
